## 03 - Conv. NN (JWE)

```python
# import modules
import matplotlib.pyplot as plt
import torch
import torchvision
import torchvision.transforms as transforms
```

```python
# import custom moudles
from week_03.cnn import Net
from week_03.plots import show_images
```

```python
# project settings
batch_size = 256
epochs = 50
learning_rate = 0.0002
```

```python
# init img transformer
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
```

```python
# load training set
trainset = torchvision.datasets.CIFAR10(root='./week_03/data', train=True, download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=batch_size, shuffle=True, num_workers=2)
```

```python
# load testset
testset = torchvision.datasets.CIFAR10(root='./week_03/data', train=False, download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=batch_size, shuffle=False, num_workers=2)
```

```python
# get some random training images and visualize them
dataiter = iter(trainloader)
images, labels = next(dataiter)
show_images(images)
```

```python
# create instance off cnn class
CNN = Net(learning_rate)
```

```python
# train the network
CNN.train(trainloader, testloader, epochs)
```

```python
# plot loss fcns
CNN.plot_loss()
```

```python
# test the network
CNN.test(testloader)
```
