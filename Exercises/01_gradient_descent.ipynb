## Week 01 - Gradient Descent (JWE)

```python
# import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import custom modules
from week_01.gradient_descent import GradientDescent
```

```python
# define target function
m = 2
b = 6
x = np.linspace(0, 1, 1000)
y = 2 * np.exp(m * x) + b
```

```python
# visualize data
plt.plot(x, y)
plt.grid()
plt.xlabel('Parameter x')
plt.ylabel('Funciton value y')
plt.show()
```

```python
# crete instance of class GradientDescent
GD = GradientDescent()
```

```python
# "train" parameter weights
GD.train(x=x, y=y, epochs=1000, learning_rate=0.01)
```

```python
# plot history
GD.plot_history(m, b)
```

```python
# plot losses
GD.plot_losses()
```
