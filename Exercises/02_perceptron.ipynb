## Week 02 - Perceptron (JWE)

```python
# import modules
import numpy as np
import matplotlib.pyplot as plt
from random import shuffle
```

```python
# import custom modules
from week_02.perceptron import Perceptron
from week_02.fcns import fy
```

```python
# file settings
size = 2000
m = 10
b = 5
epochs = 1000
learning_rate = 0.00001
```

```python
# create training data
shuffler = shuffle([i for i in range(size)])
x = np.linspace(0, 10, size)[shuffler]
y, labels = fy(x, m, b, size)
```

```python
# plot target values
plt.plot(x, y, 'r.')
plt.grid()
plt.show()
```

```python
# make ready for training
inputs = np.vstack((x, y))
```

```python
# split data (train-test-split)
split = int(0.7 * size)
inputs_train = inputs[:, 0:split].T
labels_train = labels[0:split].T
inputs_test = inputs[:, split:].T
labels_test = labels[split:].T
```

```python
# train perceptron
PCT = Perceptron(dim_inputs=2, epochs=epochs, learning_rate=learning_rate)
PCT.train(inputs_train, labels_train, inputs_test, labels_test)
labels_predict = PCT.predict_batch(inputs_test)
```

```python
# plot loss functions
PCT.plot_losses()
```
